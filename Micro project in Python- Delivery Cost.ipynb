```python
# This program calculate the delivery cost of a post company with freelance delivery men, 
#the cheapest delivery man, and total revenue.

from math import sin, cos, atan2, sqrt, radians 
import geopy.distance

#CALCULATE THE DELIVERY DISTANCE ------------------------------------------------

#Path1
from_coords = (52.086432, 4.313560)
to_coords = (52.007618, 4.357162)

#path2
#from_coords = (52.2296756, 21.0122287)
#to_coords = (52.406374, 16.9251681)

from_lat_d = from_coords[0]
from_long_d = from_coords[1]
to_lat_d = to_coords[0]
to_long_d = to_coords[1]

# calculate distance directly by calculting it from lattitude and longtitude 
def get_distance(from_lat_d, from_long_d, to_lat_d, to_long_d):
  from_lat = radians(from_lat_d)
  from_long = radians(from_long_d)
  to_lat = radians(to_lat_d)
  to_long = radians(to_long_d)
  
  dlon = to_long - from_long
  dlat = to_lat - from_lat
  
  # approximate radius of earth in km
  R = 6373.0  
  
  a = (sin(dlat/2)) ** 2 + cos(from_lat) * cos(to_lat) * (sin(dlon/2)) ** 2
  c = 2 * atan2(sqrt(a), sqrt(1-a))
  distance = R * c
  return distance 

print("The delivery distance is " + str(get_distance(from_lat_d, from_long_d, to_lat_d, to_long_d)) + " km")

#alternative way to calculate distance 
#d = geopy.distance.vincenty(from_coords, to_coords).km
#print(d)


#CALCULATE THE DELIVERY COST-------------------------------------------

SHIPPING_PRICES = {
  'Ground': 0.1,
  'Priority': 0.16,
  'Overnight': 0.23,
}

# Define calculate_shipping_cost() here:
def calculate_shipping_cost(from_coords, to_coords, shipping_type = 'Overnight'):
  #(from_lat_d, from_long_d) = from_coords 
  #(to_lat_d, to_long_d) = to_coords 
  from_lat_d = from_coords[0]
  from_long_d = from_coords[1]
  to_lat_d = to_coords[0]
  to_long_d = to_coords[1]
  distance = get_distance(from_lat_d, from_long_d, to_lat_d, to_long_d)
  shipping_rate = SHIPPING_PRICES[shipping_type]
  price = distance * shipping_rate
  return "€{0:.2f}".format(price)

# Test the function by calling 
#test_function(calculate_shipping_cost)


print("The delivery cost is " + str(calculate_shipping_cost(from_coords, to_coords, shipping_type = 'Overnight')))

#------------------------------------------------
class Driver:
  def __init__(self, speed, salary):
    self.speed = speed
    self.salary = salary

  def __repr__(self):
    return "Nile Driver with speed {} salary {} is the cheapest".format(self.speed, self.salary)

driver1 = Driver(2, 10)
driver2 = Driver(7, 20)

class Trip:
  def __init__(self, cost, driver, driver_cost):
    self.cost = cost
    driver.cost = driver_cost
    self.driver = driver

trip1 = Trip(200, driver1, 15)
trip2 = Trip(300, driver2, 40)

distance = get_distance(from_lat_d, from_long_d, to_lat_d, to_long_d)

# find the cheapest driver and cheapest driver cost 
def calculate_driver_cost(distance, *drivers):
  cheapest_driver = None
  cheapest_driver_price = None 
  for driver in drivers:
    driver_time = driver.speed * distance
    price_for_driver = driver.salary * driver_time 
    if cheapest_driver_price is None:
      cheapest_driver = driver 
      cheapest_driver_price = price_for_driver
    elif price_for_driver < cheapest_driver_price:
      cheapest_driver = driver 
      cheapest_driver_price = price_for_driver
  return "€{0:.2f}".format(cheapest_driver_price), cheapest_driver 
      

a, b = calculate_driver_cost(distance, driver1, driver2)
print("cheapest driver price is "+ str(a))
print(str(b))

# Calculate total money made 
def calculate_money_made(**trips):
  total_money_made = 0
  for trip_id, trip in trips.items():
    trip_revenue = trip.cost - trip.driver.cost
    total_money_made += trip_revenue
  return "€{0:.2f}".format(total_money_made)   
 
print("Total money made is " + calculate_money_made(trip1 = Trip(200, driver1, 15), trip2 = Trip(300, driver2, 40)))
```
